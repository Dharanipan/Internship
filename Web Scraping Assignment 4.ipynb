```python
import pandas as pd
from bs4 import BeautifulSoup
import selenium
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
driver1=webdriver.Chrome(r'chromedriver.exe')
driver1.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
rank1=[]
ranktags1=driver1.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
for i in ranktags1[:30]:
    rank1.append(i.text)
```

```python
name1=[]
nametags1=driver1.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
for i in nametags1[:30]:
    name1.append(i.text.split('[')[0])
    
```

```python
artist1=[]
artisttags1=driver1.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
for i in artisttags1[:30]:
    artist1.append(i.text)
```

```python
upload_date1=[]
uploadtags1=driver1.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
for i in uploadtags1[:30]:
    upload_date1.append(i.text)
```

```python
views1=[]
viewstags1=driver1.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
for i in viewstags1[:30]:
    views1.append(i.text)
```

```python
print(len(rank1),len(name1),len(artist1),len(upload_date1),len(views1))
```

```python
df1=pd.DataFrame({'Rank':rank1,'Name':name1,'Artist':artist1,'Upload Date':upload_date1,'Views':views1})
df1
```

```python
driver1.close()
```

```python
# 7. 
driver7=webdriver.Chrome(r'chromedriver.exe')
driver7.get("https://www.imdb.com/list/ls095964455/")
```

```python
name7=[]
year_span=[]
Genre=[]
runtime=[]
ratings=[]
votes=[]

for i in driver7.find_elements(By.XPATH,'//h3[@class="lister-item-header"]/a'):
    name7.append(i.text)

for i in driver7.find_elements(By.XPATH,'//h3[@class="lister-item-header"]/span[2]'):
    year_span.append(i.text.replace('(','').replace(')',''))
    
for i in driver7.find_elements(By.XPATH,'//span[@class="genre"]'):
    Genre.append(i.text)

for i in driver7.find_elements(By.XPATH,'//span[@class="runtime"]'):
    runtime.append(i.text)

for i in driver7.find_elements(By.XPATH,'//div[@class="ipl-rating-widget"]/div/span[2]'):
    ratings.append(i.text)

for i in driver7.find_elements(By.XPATH,'//div[@class="lister-item-content"]/p[4]/span[2]'):
    votes.append(i.text)
```

```python
df7=pd.DataFrame({'Name':name7,'Year Span':year_span,'Genre':Genre,'Runtime':runtime,'Ratings':ratings,'Votes':votes})
df7
```
