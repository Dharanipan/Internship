```python
# 1. Write a python program which searches all the product under a particular product from www.amazon.in. The 
# product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for 
# guitars.
driver1=webdriver.Chrome(r"chromedriver.exe")
driver1.get("https://www.amazon.in/")
```

```python
brand_name4=[]
smartphone_name4=[]
colour4=[]
ram4=[]
storage_rom4=[]
primary_camera4=[]
secondary_camera4=[]
display_size4=[]
battery_capacity4=[]
prize4=[]

for i in product_url4:
    driver4.get(i)
    time.sleep(5)
    try:
        bn4=driver4.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[7]/div[4]/div/div[2]/div/div[1]/table/tbody/tr[3]/td[2]/ul/li')
        brand_name4.append(bn4.text)
    except:
        brand_name4.append("--")
    try:
        spn4=driver4.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[2]/div/div[1]/h1/span')
        smartphone_name4.append(spn4.text)
    except:
        smartphone_name4.append("--")
    try:
        clr4=driver4.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[7]/div[4]/div/div[2]/div/div[1]/table/tbody/tr[4]/td[2]/ul/li')
        colour4.append(clr4.text)
    except:
        colour4.append("--")
    try:
        rm4=driver4.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[6]/div[1]/div/div[2]/ul/li[1]')
        ram4.append(rm4.text.split('|')[0])
    except:
        ram4.append("--")   
    try:
        str4=driver4.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[6]/div[1]/div/div[2]/ul/li[1]')
        storage_rom4.append(str4.text.split('|')[1])
    except:
        storage_rom4.append("--")
    try:
        pc4=driver4.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[6]/div[1]/div/div[2]/ul/li[3]')
        primary_camera4.append(pc4.text)
    except:
        primary_camera4.append("--")
    try:
        sc4=driver4.find_element(By.XPATH,'')
        secondary_camera4.append(sc4.text)
    except:
        secondary_camera4.append("--")
    try:
        ds4=driver4.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[6]/div[1]/div/div[2]/ul/li[2]')
        display_size4.append(ds4.text)
    except:
        display_size4.append("--")
    try:
        bc4=driver4.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[6]/div[1]/div/div[2]/ul/li[4]')
        battery_capacity4.append(bc4.text)
    except:
        battery_capacity4.append("--")
    try:
        prz4=driver4.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[2]/div/div[3]/div[1]/div/div[1]')
        prize4.append(prz4.text)
    except:
        prize4.append("--")
        
```

```python
df4=pd.DataFrame({'Brand Name':brand_name4,'Smart Name':smartphone_name4,'Colour':colour4,'Ram':ram4,'Storage Rom':storage_rom4,'Primary Camera':primary_camera4,'Display Size':display_size4,'Battery Capacity':battery_capacity4,'Prize':prize4,'Product URL':product_url4})
df4
```
